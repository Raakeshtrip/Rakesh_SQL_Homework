```python
employeeDetailsQuery = "select title,avg(salary) as Salary_by_Title from titles e JOIN salary s ON s.emp_no=e.emp_no group by title"
```

```python
df_1 = pd.read_sql_query(employeeDetailsQuery,con=connection)
df_1.set_index("title",inplace=True)
df_1.head()

df_1.iloc[::-1].plot.barh(title="Salary Avg")
plt.tight_layout()
plt.show()
```
